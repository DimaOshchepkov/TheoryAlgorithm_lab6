```python
import random
import matplotlib.pyplot as plt
import time
```

```python
def linear_search(arr, target):

    index = 0
    while index < len(arr) and arr[index] != target:
        index += 1
    return arr[index] if index < len(arr) else -1

def linear_search_wall(arr, target):

    index = 0
    arr.append(target)
    while arr[index] != target:
        index += 1
    arr.pop()
    return arr[index] if index < len(arr) else -1

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        # Флаг, показывающий, были ли сделаны перестановки на текущем проходе
        swapped = False
        for j in range(0, n-i-1):
            # Сравниваем два соседних элемента
            if arr[j] > arr[j+1]:
                # Если текущий элемент больше следующего, меняем их местами
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        # Если на текущем проходе не было сделано ни одной перестановки, значит, список уже отсортирован
        if not swapped:
            break

def selection_sort(arr):

    n = len(arr)
    for i in range(n):
        # Поиск индекса минимального элемента в оставшейся части списка
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        # Обмен текущего элемента с минимальным элементом
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

def generate_random_array(length, min_value=0, max_value=100):
    return [random.randint(min_value, max_value) for _ in range(length)]
```

```python
# Параметры для построения графика
array_lengths = [i for i in range(10_000, 1_000_000, 10_000)]  # Длины массивов
execution_times = []  # Список времен выполнения для каждой длины массива

# Измеряем время выполнения линейного поиска для каждой длины массива
for length in array_lengths:
    arr = generate_random_array(length)
    search_el = 500
    arr[int(len(arr)/2)] = search_el
    start_time = time.time()
    linear_search(arr, search_el)  # Поиск случайного элемента
    end_time = time.time()
    execution_times.append(end_time - start_time)
    del arr

# Построение графика
plt.plot(array_lengths, execution_times, marker='o')
plt.title('Временная сложность линейного поиска d среднем')
plt.xlabel('Длина массива')
plt.ylabel('Время выполнения (сек)')
plt.grid(True)
plt.show()
```
